```python
# Handling missing values by replacing Null or Nan string value in the data with an actual NAN recognized by python using np.nan
df_copy.replace(["NULL", "NaN"], np.nan, inplace=True)
```

```python
# Only Medication_adherence have missing values of 3,131
# Since Medication_adherence is a categorical column
df.fillna(df.mode().iloc[0], inplace=True)

# Fill in the missing values for Medication_adherence with
df.replace(["Undiagnosed"], pd.NA, inplace=True)
```

```python
# Dropping critical values/columns
df_copy = df_copy.drop(columns=["patient_id"])
```

```python
# Fixing data types


# Numeric columns
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["bmi"] = pd.to_numeric(df["bmi"], errors="coerce")
df["systolic_bp_mmhg"] = pd.to_numeric(df["systolic_bp_mmhg"], errors="coerce")
df["diastolic_bp_mmhg"] = pd.to_numeric(df["diastolic_bp_mmhg"], errors="coerce")
df["total_cholesterol_mg_dl"] = pd.to_numeric(df["total_cholesterol_mg_dl"], errors="coerce")
df["ldl_mg_dl"] = pd.to_numeric(df["ldl_mg_dl"], errors="coerce")
df["hdl_mg_dl"] = pd.to_numeric(df["hdl_mg_dl"], errors="coerce")
df["creatinine_mg_dl"] = pd.to_numeric(df["creatinine_mg_dl"], errors="coerce")

# Categorical columns
df["sex"] = df["sex"].astype("category")
df["residence"] = df["residence"].astype("category")
df["family_history_hypertension"] = df["family_history_hypertension"].astype("category")
df["diabetes"] = df["diabetes"].astype("category")
df["smoking"] = df["smoking"].astype("category")
df["alcohol_heavy"] = df["alcohol_heavy"].astype("category")
df["physically_active"] = df["physically_active"].astype("category")
df["high_salt_diet"] = df["high_salt_diet"].astype("category")
df["hypertension_status"] = df["hypertension_status"].astype("category")
df["diagnosed"] = df["diagnosed"].astype("category")
df["on_medication"] = df["on_medication"].astype("category")
df["medication_adherence"] = df["medication_adherence"].astype("category")
df["bp_controlled"] = df["bp_controlled"].astype("category")
df["stroke_history"] = df["stroke_history"].astype("category")
df["myocardial_infarction"] = df["myocardial_infarction"].astype("category")
df["heart_failure"] = df["heart_failure"].astype("category")

print(df.dtypes)
```

```python
# using the interQuartile Range (IQR) to identify outliers
Q1 = df_copy["bmi"].quantile(0.25)
Q3 = df_copy["bmi"].quantile(0.75)
IQR = Q3 - Q1

lowerbound = Q1 - 1.5 * IQR
upperbound = Q3 + 1.5 * IQR

outlier = df_copy[(df_copy["bmi"] < lowerbound) | (df_copy["bmi"] > upperbound)]

# capping out outliers in bmi
df_copy["bmi"] = df_copy["bmi"].clip(lowerbound, upperbound)
```
